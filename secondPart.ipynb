```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
raw = pd.read_csv("data.csv")
filter = [col for col in raw.columns if (not col in ["Sample ID", "Class"]) ]
df = raw[filter].replace(np.NaN, 0)
for i in df.columns:
    if(np.sum(df[i]) == 0):
        del df[i]
corrF = df.corr()
plt.matshow(corrF)
# https://stackoverflow.com/questions/29432629/plot-correlation-matrix-using-pandas
import matplotlib

matplotlib.rc('figure', figsize=(10, 5))
plt.xticks([i for i in range(len(corrF.columns))], list(corrF.columns), rotation=45)
null = plt.yticks([i for i in range(len(corrF.columns))], list(corrF.columns))
```
