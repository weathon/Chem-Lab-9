```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
raw = pd.read_csv("data.csv")
filter = [col for col in raw.columns if (not col in ["Sample ID", "Class"]) ]
df = raw[filter].replace(np.NaN, 0)
for i in df.columns:
    if(np.sum(df[i]) == 0):
        del df[i]
corrF = df.corr()
plt.matshow(corrF)
plt.xticks([i for i in range(len(corrF.columns))], list(corrF.columns), rotation=90)
null = plt.yticks([i for i in range(len(corrF.columns))], list(corrF.columns))
```

```python
corrA = np.array(corrF)
corrA[(corrA < 0.8) & (corrA > -0.8)] = 0
corrA[(corrA > 0.8)] = 0.1
corrA[(corrA < -0.8)] = 1
plt.xlabel("")
plt.matshow(corrA, cmap="BrBG")
```
